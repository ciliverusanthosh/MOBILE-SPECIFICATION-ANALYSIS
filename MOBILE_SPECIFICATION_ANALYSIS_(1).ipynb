```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Mobile-Price.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

DISTRIBUTION OF RAM SIZE UNDER  PRICE 20000


```python
filtered_df = df[(df['Price'] >= 0) & (df['Price'] <= 20000)]

plt.scatter(filtered_df['Price'],filtered_df['RAM'])
plt.xlabel('Price')
plt.ylabel('RAM')
plt.title('Price vs RAM')
plt.show()
```

DISTRIBUTOIN OF BATTERY SIZE UNDER PRICE 20000


```python
filtered_df = df[(df['Price'] >= 0) & (df['Price'] <= 20000)]
plt.figure(figsize=(10,5))
plt.scatter(filtered_df['Price'],filtered_df['Battery_Power'])
plt.xlabel('Price')
plt.ylabel('Battery_Power')
plt.title('Price vs Battery_Power')
plt.show()
```

```python
filtered_df = df[(df['Price'] >= 0) & (df['Price'] <= 20000)]
ax = sns.histplot(filtered_df['Price'],kde=True,color='Skyblue')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
corr = df.corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='RAM', y='Price', data=df)
plt.xlabel('RAM')
plt.ylabel('Price')
plt.title('Box Plot of Price by RAM')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x='ROM', y='Price', data=df)
plt.xlabel('ROM')
plt.ylabel('Price')
plt.title('Box Plot of Price by ROM')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Price', y='Primary_Cam')
plt.xlabel('Price')
plt.ylabel('Primary Camera (MP)')
plt.title('Price vs Primary Camera')
plt.show()


plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Price', y='Selfi_Cam')
plt.xlabel('Price')
plt.ylabel('Selfie Camera (MP)')
plt.title('Price vs Selfie Camera')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Price', y='Mobile_Size')
plt.xlabel('Price')
plt.ylabel('Mobile Size')
plt.title('Price vs Mobile Size')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Ratings'], kde=True)
plt.xlabel('Ratings')
plt.title('Distribution of Ratings')
plt.show()
```

```python
features = ['RAM', 'ROM', 'Mobile_Size', 'Battery_Power', 'Primary_Cam', 'Selfi_Cam']
for i in features:
    plt.figure(figsize=(10, 5))
    sns.scatterplot(data=df, x='Price', y=i)
    plt.xlabel('Price')
    plt.ylabel(i)
    plt.title(f'Price vs {i}')
    plt.show()
```
